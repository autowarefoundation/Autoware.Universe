## Logistic function

```python
time = np.linspace(0, 50, 501)
```

```python
time
```

```python
def logistic_map(t):
    L = 10 # maximum y
    k=0.2    # slope 
    x0 = 25 # where x0 starts
    
    return L / (1 + np.exp(-k*(t-x0))) 
```

```python
y = logistic_map(time)
```

```python
plt.plot(time, y )
plt.grid()
```
